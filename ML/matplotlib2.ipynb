```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.scatter(x = iris['sepal_length'], y = iris['petal_length'], c = iris['species'], cmap ='jet', alpha = 0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,8), nrows = 2, ncols=1, sharex=True)
ax[0].scatter(df['avg'], df['strike_rate'], color = 'red', marker = '*')
ax[1].scatter(df['avg'], df['runs'], color = "green", marker = '.')

ax[0].set_title("Avg vs Strike Rate")
ax[1].set_title("Avg vs Runs")

ax[0].set_ylabel("Strike Rate")
ax[1].set_ylabel("Runs")
ax[1].set_xlabel("Avg")
# ax[0].set_xlabel("avg")
```

```python
## 3D surface plots

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
print(x)
```

```python
xx,yy = np.meshgrid(x,y)
print(xx.shape)
print(yy.shape)
```

```python
z = xx ** 2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize= (12,8))
ax = plt.subplot(projection = '3d')
ax.plot_surface(xx,yy,z, cmap = 'viridis')
```
