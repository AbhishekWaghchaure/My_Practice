```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
X,y = make_classification(n_samples=10000,n_classes=2,n_features=3,random_state=32,n_redundant=1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=32)
```

```python
model1 = KNeighborsClassifier(algorithm='auto', n_neighbors=5)
```

```python
model1.fit(X_train,y_train)
```

```python
model1_y_pred = model1.predict(X_test)
```

```python
cf = confusion_matrix(y_test,model1_y_pred)
sns.heatmap(cf,annot =True,fmt='d')
```

```python
print(classification_report(y_test,model1_y_pred))
```
