```python
from torch.utils.data import DataLoader, TensorDataset
import torch.nn.functional as F
```

```python
train_dataset_features = torch.tensor(X_train.to_numpy())
train_dataset_targets = torch.tensor(y_train.to_numpy())

## One hot Encoding for targets
train_dataset_targets =F.one_hot(train_dataset_targets, num_classes=9)

training_data = TensorDataset(train_dataset_features,train_dataset_targets)
print(training_data)

next(iter(training_data))
```

```python
test_dataset_features = torch.tensor(X_test.to_numpy())
test_dataset_targets = torch.tensor(y_test.to_numpy())

## One Hot Encoding for test Targets
test_dataset_targets = F.one_hot(test_dataset_targets, num_classes=9)

testing_data = TensorDataset(test_dataset_features,test_dataset_targets)
```

```python
print(train_dataset_features.shape)
print(train_dataset_targets.shape)
print(test_dataset_features.shape)
print(test_dataset_targets.shape)
```

```python
dataloader = DataLoader(training_data,batch_size=32,shuffle=True)
```

```python
for features, label in dataloader:
    print(features, label)
```

```python
# model = nn.Sequential(
#     nn.Linear(11,30),
#     nn.Linear(30,12),
#     nn.Linear(12,9),
#     nn.Sigmoid()
# )
```

```python
class model(nn.Module):
    def __init__(self):
        super(model,self).__init__()
        self.fc1 = nn.Linear(11,30)
        self.fc2 = nn.Linear(30,12)
        self.fc3 = nn.Linear(12,9)
        
    def forward(self,x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.softmax(self.fc3(x))
        return x
        
```

```python
model = model()
```

```python
from torch.optim import SGD
```

```python
criterian = nn.CrossEntropyLoss()
optimizer = SGD(model.parameters(),lr =0.96)
```

```python
model.parameters
```

```python
## training Loop
for features, labels in dataloader:
    for epoch in range(11):
        optimizer.zero_grad()
        output = model(features.float())
        loss = criterian(output,labels.float())
        loss.backward()
        optimizer.step()
        
        print(f'Epoch {epoch+1}, Loss: {loss.item()}')
```
